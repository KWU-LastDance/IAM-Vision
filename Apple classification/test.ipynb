```python
from torch.utils.data import DataLoader
import torch
import torch.nn as nn
import numpy as np
from datas.dataset import make_dataset
import torch.optim as optim
from datas.config import cfg
from models.apple_model import Efficientnet_AppleClassifier, Densenet_AppleClassifier, Resnet_AppleClassifier
from tests.test_model import test_and_visualize
```

```python
image_dir_test = './datas/test/'
csv_dir_test = './datas/test/'
image_size=cfg['image_size']
test_set = make_dataset(
    image_dir=image_dir_test,
    csv_dir=csv_dir_test,
    image_size=image_size
)


test_loader = DataLoader(
    test_set, 
    batch_size=cfg['batch_size'],
    shuffle=True
)
```

```python

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = Densenet_AppleClassifier(cfg).to(device)
optimizer = optim.Adam(model.parameters(), lr=cfg['learning_rate'], weight_decay=cfg['weight_decay'])  # L2 정규화 적용

# 손실 함수 설정
criterion = nn.CrossEntropyLoss()
```

```python

# 모델 상태 로드
def load_model(model, load_path, map_location=None):
    model.load_state_dict(torch.load(load_path, map_location=map_location))
    return model

# 저장된 모델 상태를 불러오는 코드
load_path = './weight/densenet_best_model.pth'
model = load_model(model, load_path, map_location=torch.device('cpu'))
#model = load_model(model, load_path)
```

```python
# 모델 검증 및 시각화 실행
test_and_visualize(model, test_loader, criterion, device)
```
